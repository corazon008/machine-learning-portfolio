```python
from reinforcement_learning.game.Party import Party
from reinforcement_learning.players import RealPlayer, RandomPlayer, QLearningPlayer
```

```python
player1 = QLearningPlayer.QLearningPlayer(1)
player2 = RandomPlayer.RandomPlayer(-1)
party = Party(player1, player2)
```

```python
epochs = 101000
wins = 0
win_rates = []
```

```python
for i in range(epochs):
    party.play()

    if party.winner == player1:
        wins += 1
    win_rates.append(wins/(1+len(win_rates)))
    print(f"\rwinrate: {win_rates[-1]}", end="")

    party.reset_game()
```

```python
import seaborn as sns

sns.lineplot(x=range(len(win_rates)), y=win_rates)
```
